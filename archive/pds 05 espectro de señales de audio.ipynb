```python
def plot_spectrum(s, Fs):
    S = fft(s) / len(s)
    S = fftshift(S)
    f = np.linspace(-Fs/2, Fs/2, len(S))

    S_mag = np.abs(S)
    S_mag_dB = 10 * np.log(S_mag)

    plt.figure()
    plt.plot(f, S_mag)
    plt.title("Espectro de la señal")

    
    plt.figure()
    plt.plot(f, S_mag_dB)
    plt.title("Espectro de la señal (dB)")


    plt.figure()
    plt.specgram(s, Fs=Fs)
    plt.title("Espectrograma de la señal")

```

```python
Fs = 44100 
ts = 1/Fs

T = 10
N = int(T/ts)

t = np.linspace(0, T, N)
s = np.cos(2*np.pi*400*t)

Audio(s, rate=Fs)
```

```python
plot_spectrum(s, Fs)
```

```python
Fs = 44100 
ts = 1/Fs

T = 10
N = int(T/ts)

t = np.linspace(0, T, N)
s = np.cos(2*np.pi*350*t) + np.cos(2*np.pi*440*t)

Audio(s, rate=Fs)
plot_spectrum(s, Fs)
```

```python
Fs = 44100 
ts = 1/Fs

T = 16
N = int(T/ts)

t = np.linspace(0, T, N)

f = lambda x: np.cos(2*np.pi*480*x) + np.cos(2*np.pi*440*x)

s = np.piecewise(t,
    [
        ((t > 0) & (t <= 2)),
     ((t > 2) & (t <= 6)),
     ((t > 6) & (t <= 8)),
     ((t > 12) & (t <= 16)),
     ],
    [f, 0,f, 0])

Audio(s, rate=Fs)
plot_spectrum(s, Fs)
```
