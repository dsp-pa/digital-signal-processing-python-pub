```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import fft

T = 50
Fs = 5

Ts = 1 / Fs
N = int(T / Ts)

t = np.arange(0, T, Ts)
s_t = np.cos(2*np.pi*t) + np.cos(2* 2* np.pi*t)

S_f = fft.fft(s_t) / (T / Ts)

plt.figure(figsize=(8,6), dpi=150)
plt.plot(s_t)


```

```python

S = fft.fftshift(S_f)

Fs = N / T

f1 = np.linspace(0, Fs, N)
f2 = np.linspace(-Fs/2, Fs/2, N)

plt.figure()

plt.figure(figsize=(8,6), dpi=150)
plt.plot(f1, np.abs(S_f))

plt.figure()
plt.figure(figsize=(8,6), dpi=150)
plt.plot(f2, abs(S))
```
